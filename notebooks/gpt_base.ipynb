```python
import glob
```

```python
dirs = {
    'optc_s': '../results/gpt_cls/optc/results_snapshot_static*',
    'optc_t': '../results/gpt_cls/optc/results_snapshot_bert*',
    'unsw_t': '../results/gpt/unsw/snapshot-ft_results_snapshot_bert_static_*',
    'unsw_s': '../results/gpt/unsw/snapshot-ft_results_snapshot_bert_tiny_*',
    'lanl_s': '../results/gpt/lanl14argus/*_bert_static*.txt',
    'lanl_t': '../results/gpt/lanl14argus/*_bert_temporal*.txt'
}

files = dict()
for k,d in dirs.items(): 
    results = glob.glob(d)
    files[k] = results 

files
```

```python
best = dict()
for ds,fs in files.items(): 
    best[ds] = (0,0,0,0)
    
    for f in fs: 
        with open(f, 'r') as f_: 
            raw = f_.read()
        
        raw = raw.split('\n')
        raw = [[float(val) for val in r.split(',')] for r in raw[2:-1]]
        
        if not raw: 
            continue 
        
        best_record = max(raw, key=lambda x : x[4])[2:]

        if best_record[1] > best[ds][1]: 
            best[ds] = best_record 

best
```
